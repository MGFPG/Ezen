```python
#국영수 세과목 점수를 딕셔너리에 입력 합과 평균을 출력하라

st = dict()

st['kor'] = int(input('국어 : '))
st['eng'] = int(input('영어 : '))
st['mat'] = int(input('수학 : '))
st['tot'] = st['kor'] + st['eng'] + st['mat']
st['ave'] = st['tot'] / 3

print("국어 영어 수학 총점 평균")

print("%3d %3d  %3d  %3d %6.2f"% (st['kor'], st['eng'], st['mat'],st['tot'], st['ave']))
```

```python
def cal(v1,v2,op):
    result = 0
    
    if op == '+':
        result=v1+v2
        print(v1,'+',v2)
    elif op == '-':
        result=v1-v2
        print(v1,'-',v2)
    elif op == '*':
        result=v1*v2
        print(v1,'*',v2)
    elif op == '/':
        result=v1/v2
        print(v1,'/',v2)
    elif op == '%':
        result=v1%v2
        print(v1,'%',v2)
        
    else:
        print('Error opration')
        
    return result

print('+ simple')
r=cal(10,20,'+')
print('result:',r)

print('- simple')
r=cal(10,20,'-')
print('result:',r)

print('* simple')
r=cal(10,20,'*')
print('result:',r)

print('/ simple')
r=cal(10,20,'/')
print('result:',r)

print('% simple')
r=cal(10,20,'%')
print('result:',r)
```

```python
# lotto 추첨기
#함수로 구현하기
#추첨하는 개수를 입력하시요. 6자릿수 입력 리스트
# 00~30 사이의 수가 나오게 하기 랜덤함수

import random

def Lotto():
    hit = 6
    lotto_num = list()

    while hit > 0:
        num = random.randint(0,30)
        if num in lotto_num:
            continue
        else:
            lotto_num.append(num)
            hit -= 1

    return sorted(lotto_num)


print(Lotto())
```
